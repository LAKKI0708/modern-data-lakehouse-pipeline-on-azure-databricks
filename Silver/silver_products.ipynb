```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
from pyspark import *
```

```python
df_products = spark.read.table("advenworks_cata.bronze.products")
df_products = df_products.withColumnRenamed("rowguid", "Product_rowguid")
df_products = df_products.withColumnRenamed("Name", "Product_Name")
df_products = df_products.withColumnRenamed("SellStartDate", "Product_SellStartDate")
df_products = df_products.withColumnRenamed("SellEndDate", "Product_SellEndDate")
df_products = df_products.withColumnRenamed("DiscontinuedDate", "Product_DiscontinuedDate")
df_products = df_products.withColumnRenamed("ModifiedDate", "Product_ModifiedDate")
df_products = df_products.withColumn("Product_SellStartDate", to_timestamp(col('Product_SellStartDate')))
df_products = df_products.withColumn("Product_SellEndDate", to_timestamp(col('Product_SellEndDate')))
df_products = df_products.withColumn("Product_ModifiedDate", to_timestamp(col('Product_ModifiedDate')))
df_products = df_products.withColumn("Product_DiscontinuedDate", to_timestamp(col('Product_DiscontinuedDate')))
df_products = df_products.drop("_rescued_data")
df_products.display()
```

```python
df_productcategory = spark.read.table("advenworks_cata.bronze.productcategory")
df_productcategory = df_productcategory.withColumnRenamed("Name", "Productcategory_Name")
df_productcategory = df_productcategory.withColumnRenamed("rowguid", "Productcategory_rowguid")
df_productcategory = df_productcategory.withColumnRenamed("ModifiedDate", "Productcategory_ModifiedDate")
df_productcategory = df_productcategory.withColumn("Productcategory_ModifiedDate", to_timestamp(col('Productcategory_ModifiedDate')))
df_productcategory = df_productcategory.drop("_rescued_data")
df_productcategory.display()
```

```python
df_productsubcategory = spark.read.table("advenworks_cata.bronze.productsubcategory")
df_productsubcategory = df_productsubcategory.withColumnRenamed("Name", "Productsubcategory_Name")
df_productsubcategory = df_productsubcategory.withColumnRenamed("rowguid", "Productsubcategory_rowguid")
df_productsubcategory = df_productsubcategory.withColumnRenamed("ModifiedDate", "Productsubcategory_ModifiedDate")
df_productsubcategory = df_productsubcategory.withColumn("Productsubcategory_ModifiedDate", to_timestamp(col('Productsubcategory_ModifiedDate')))
df_productsubcategory = df_productsubcategory.drop("_rescued_data")
df_productsubcategory.display()
```

# Joining productcategory and productsubcatory tables on products table

```python
df_product_with_subcat = df_products.join(
    df_productsubcategory, on="ProductSubcategoryID", how="left"
)

df_product_full = df_product_with_subcat.join(
    df_productcategory, on="ProductCategoryID", how="left"
)
```

```python
df_product_full = df_product_full.withColumn(
    "IsZeroPricing",
    (col("StandardCost") == 0) | (col("ListPrice") == 0)
)

df_product_full.display()
```

```python
df_product_full.write.format("delta")\
    .mode("overwrite")\
    .option("path","abfss://silver@advenworksdl.dfs.core.windows.net/products")\
    .save()
```
